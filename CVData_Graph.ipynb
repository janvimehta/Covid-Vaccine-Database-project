```python
from scipy import stats
r,p =stats.pearsonr(df.Total_Individuals_Vaccinated, df.Male_Doses_Administered)
print(round(r,3))
print(round(p,20))
```

```python
corr_df = pd.DataFrame(columns=['r','p'])

for col in df:
    if pd.api.types.is_numeric_dtype(df[col]) and col != 'Total_Individuals_Vaccinated':
        r, p = stats.pearsonr(df.Total_Individuals_Vaccinated, df[col])
        corr_df.loc[col] = [round(r, 3), round(p, 20)]
        
corr_df
```

```python
import matplotlib.pyplot as plt
plt.scatter(df.Female_Doses_Administered, df.Total_Doses_Administered)
plt.title('Vaccination Graph')
plt.xlabel('Female_Doses_Administered')
plt.ylabel('Total_Doses_Administered')
plt.show()
```

```python
df_State = df[df['State'] == 'Gujarat']
df_DiffState = df[df['State'] == 'Maharashtra']

plt.scatter(df_State.Female_Doses_Administered, df_State.Total_Doses_Administered, label='Gujarat')
plt.scatter(df_DiffState.Female_Doses_Administered, df_DiffState.Total_Doses_Administered, label='Maharashtra')
plt.title('Vaccination Graph')
plt.xlabel('Female_Doses_Administered')
plt.ylabel('Total_Doses_Administered')
plt.legend()
plt.show()
```

```python
df_State_reduced = df_State.sample(50)
df_DiffState_reduced = df_DiffState.sample(50)

plt.scatter(df_State_reduced.Female_Doses_Administered, df_State_reduced.Total_Doses_Administered, label='Gujarat')
plt.scatter(df_DiffState_reduced.Female_Doses_Administered, df_DiffState_reduced.Total_Doses_Administered, label='Maharashtra')
plt.title('Vaccination Graph')
plt.xlabel('Female_Doses_Administered')
plt.ylabel('Total_Doses_Administered')
plt.legend()
plt.show()
```
